### Selecting the right visualization:
<b>Depends on:</b>
* Number of variables
 - 1D (Univariate)
 - 2D (Bivariate)
 - 3D+
* Type of data:
 - Discrete/Categorical (Ordinal & Nominal)
 - Continuous/Numeric (Interval & Ratio)

<b>Do you want to compare values?</b> If you need to compare one or many value sets, and need to visualize the lows and highs in the data sets, you would need:
- <b>Column Chart aka Bar Chart</b> (`plt.bar`, [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)) <u>Categorical, 1D/2D</u>: A column chart is used to show a comparison among different items. While this looks similar to a histogram, it shouldn't be used to show distribution.

- <b>Side-by-side Column Chart</b> (`plt.bar(1,h1,w), plt.bar(1+w,h2,w)`, [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)) <u>Categorical, 3D+</u>: Like a column chart, but it has multiple columns for each x. This should be used to compare items across different x values AND within the same x value.

- <b>Horizontal Bar Plot</b> (`plt.barh`, [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.barh.html)) <u>Categorical, 1D/2D</u>: A bar graph, basically a horizontal column chart, should be used to avoid clutter when one data label is long or if you have more than 10 items to compare. This type of visualization can also be used to display negative numbers.

- <b>Line Plot</b> (`plt.plot`, [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html)) <u>Numerical, 2D</u>: A line graph reveals trends or progress over time and can be used to show many different categories of data. You should use it when you chart points that are related to neighbouring points, such as a time series. It shouldn't be used for independent datapoints. Add extra lines for an extra categorical variable represented by a legend.

- <b>Scatter Plot</b> (`plt.scatter`, [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html)) <u>Numerical, 2D</u>: A scatter plot will show the relationship between two different variables or it can reveal the distribution of the y variable as the x value changes. It should be used when there are many different data points, and you want to highlight similarities in the data set. It's also useful when looking for outliers or for understanding the distribution of your data. It shouldn't be used to show the distribution if the x value is irrelevant.

<b>Do you want to show the composition of something?</b> Use this type of chart to show how individual parts make up the whole of something:
- <b>Pie Chart</b> (`plt.pie`, [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.pie.html)) <u>Categorical, 1D</u>: A pie chart shows a static percentages and how categories represent part of a whole. A pie chart represents numbers in percentages, and the total sum of all segments needs to equal 100%. Do not use this if you have more than 10 or so categories (you can combine many into an 'Other' category if this helps), or when the percentages are similar. Remember, if you exclude the percentages from your graph, does it still prove your point?

- <b>Stacked Column Chart</b> (`plt.bar(x1,h1), plt.bar(x2,h2,bottom=h1)`, [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)) <u>Categorical, 3D</u>: This should be used to compare many different items across different x values (usually year or category), and show the composition of each item being compared. This shouldn't be used if you're trying to compare items within the same x value.

- <b>Area Chart</b> (`plt.stackplot`, [documentation](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.stackplot.html)) <u>Numerical/Categorical, 1D</u>: An area chart is basically a line chart, but the space between the x-axis and the line is filled with a color or pattern. It is useful for showing part-to-whole relations. It helps you analyze both overall and individual trend information. These should be used cautiously, as it can be unclear whether the data is stacked or overlapping. Add extra lines for an extra categorical variable represented by a legend.


<b>Do you want to understand the distribution of your data?</b> Distribution charts help you to understand outliers, the normal tendency, and the range of information in your values.
- <b>Scatter Plot</b> (`plt.scatter`, [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html)) <u>Numerical, 2D</u>: In addition to the description about scatter plots earlier, scatter plots are especially useful to show the location of outliers (in terms of both x and y), and how the distribution changes as x increases.

- <b>Box Plot aka Box and Whisker Plot</b> (`plt.boxplot`, [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.boxplot.html)) <u>Categorical, 1D/2D</u>: A boxplot is a way to show the distribution of one-dimensional data based on the five-number summary: minimum, first quartile, median, third quartile, and maximum. This type of plot can also show outliers, and you can create side-by-side boxplots to compare distributions between categories (eg. height of each gender).

- <b>Histogram</b> (`plt.hist`, [documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html)) <u>Numerical, 1D</u>: A histogram is an accurate representation of the distribution of numeric data. To create a histogram, first we divide the entire range of values into a series of intervals (or 'bins'), and second we count how many values fall into each interval. The bins are consecutive and non-overlapping intervals of a variable. Can have a <b>Density Curve</b>, which is essentially a smoothed histogram.

<b>Do you want to better understand the relationship between value sets?</b> Relationship charts are suited to showing how one variable relates to one or numerous different variables. You could use this to show how something positively effects, has no effect, or negatively effects another variable.

- <b>Scatter Plot</b>, <b>Coloured Scatter Plot</b> (`plt.scatter(c=colours)`, [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html)) or <b>Bubble Chart</b> (`plt.scatter(s=size)`, [documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html)) <u>Numerical, 2D/3D+</u>: A coloured scatter or bubble chart is similar to a normal scatter plot in that it can show distribution or relationship, but it includes a third data set, which is indicated by the colour / size of the bubble. Try not to create a coloured bubble chart - that's usually too much to interpret for the viewer.

<b>Other:</b>
- <b>Dot Chart</b> <u>Categorical, 1D</u>: Similar to column chart, but with dots.

- <b>Table</b> <u>Categorical, 2D</u>: Simple table with columns and rows.

- <b>Heatmap aka Matrix</b> <u>Categorical, 2D</u>: Like a table, but with colour representing an interval.

- <b>Scatter Matrix</b> <u>Numerical, 2D/3D+ </u>: Multiple scatter plots divided into a matrix.

### Line Plot:

```python
# simple line plot
# create figure
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(221)

# create plot
categories = pd.unique(exampleData["species"])
means = []
for c in categories:
    f = exampleData["species"] == c
    means.append(exampleData.where(f)["sepal_length"].mean())
plt.plot(categories, means)

# format axis
ax1.set_xlabel("x Axis")
ax1.set_ylabel("y Axis")
ax1.set_title("Plot Title")

# display
plt.show()
```

```python
# set numOfQuan to 0 for non numetric
def genGroups(p, isNumetric, numOfQuan):
    if isNumetric:
        q = exampleData[p].quantile([(x+1)/numOfQuan for x in range(numOfQuan-1)]).values
        q = np.insert(q, 0, 0)
        q = np.append(q, exampleData[p].max()+1)
        #q = q.round().astype(int)
        return [[q[i], q[i+1]] for i in range(len(q)-1)]
    else:
        return pd.unique(exampleData[p])

# generate a filter based on a range or group
def genFilter(cond, param, isNumetric):
    if isNumetric:
        return (exampleData[param] >= cond[0]) & (exampleData[param] <= cond[1])
    else:
        return exampleData[param] == cond

# generate quantiles for frames
quan = genGroups("sepal_width", True, 5)
quanTicks = [f"{round(min(i)/100, 2)}-{round(max(i)/100, 2)}" for i in quan]

# calculate means for each frames quantile and target group
def getMeans(cond, param):
    utimeMeans = []
    for g in quan:
        f1 = genFilter(cond, param, False)
        f2 = genFilter(g, "sepal_width", True)
        utimeMeans.append(exampleData.where(f1).where(f2)["sepal_length"].mean())
    return pd.Series(utimeMeans)

# line plot with multiple lines
fig = plt.figure(figsize=(12,10))
fig.subplots_adjust(wspace=0.3, hspace=0.3)

# observe relationship for each output codec
ax1 = fig.add_subplot(111)
groups = pd.unique(exampleData["species"])

# plot line
colors1 = cm.Accent(np.arange(3)/3.)
for l, c in enumerate(groups):
    plt.plot(quanTicks, getMeans(c, "species"), alpha=0.5, color=colors1[:][l])

# legend
handles1 = [plt.Line2D([0], [0], marker='s', color='w', label='Bar',
                       markerfacecolor=colors1[:][i], markersize=10) for i in range(len(groups))]
ax1.legend(handles1, groups, frameon=True, facecolor="white", title="Groups")

# format axis
ax1.set_xlabel("x Axis")
ax1.set_ylabel("y Axis")
ax1.set_title("Plot Title")

plt.show()
```

### Side-by-Side Column Chart:

```python
# includes functions for generalization
# example groups data into quantiles and groups
fig = plt.figure(figsize=(12,10))
fig.subplots_adjust(wspace=0.3, hspace=0.3)

# function for generating groups based on quantiles
# set numOfQuan to 0 for non numetric
def genGroups(p, isNumetric, numOfQuan):
    if isNumetric:
        q = exampleData[p].quantile([(x+1)/numOfQuan for x in range(numOfQuan-1)]).values
        q = np.insert(q, 0, 0)
        q = np.append(q, exampleData[p].max()+1)
        #q = q.round().astype(int)
        return [[q[i], q[i+1]] for i in range(len(q)-1)]
    else:
        return pd.unique(exampleData[p])

# generate a filter based on a range or group
def genFilter(cond, param, isNumetric):
    if isNumetric:
        return (exampleData[param] >= cond[0]) & (exampleData[param] <= cond[1])
    else:
        return exampleData[param] == cond

def sbsColumn(ax, group, paramX, paramY):
    colors1 = cm.Accent(np.arange(4)/4.)
    width = 0.5
    isParamNumetric = pd.api.types.is_numeric_dtype(exampleData[group])
    
    params = genGroups(group, isParamNumetric, 4)
    
    # generate bins based on quantiles
    bins1 = genGroups(paramX, True, 5)

    # plot bars
    for i, r in enumerate(bins1):
        # place data into bins
        b = exampleData.where((exampleData[paramX] >= min(r)) &
                            (exampleData[paramX] <= max(r)),
                            inplace=False)
        # place data into groups
        for l, p in enumerate(params):            
            f = genFilter(p, group, isParamNumetric)
            plt.bar(1+(i*width*4.5)+(width*0.8*l), b.where(f)[paramY].mean(), width*0.8, color=colors1[:][l])

    # legend for codec groups
    handles1 = [plt.Line2D([0], [0], marker='s', color='w', label='Bar',
                              markerfacecolor=colors1[:][i], markersize=10) for i in range(len(params))]
    
    # format legend
    if isParamNumetric:
        d = 10**(len(str(max(params[0])))-2)
        if d > 1:
            t = f"{group}/{d}"
        else:
            t = group
        ax.legend(handles1, 
                  [f"{(int) (min(p)/d)}-{(int) (max(p)/d)}" for c in bins1  for i, p in enumerate(params)], 
                  frameon=True, 
                  facecolor="white",
                  title=t)
    else:
        ax.legend(handles1, params, frameon=True, facecolor="white", title=group)

    # format axis
    plt.xticks(0.8+width*0.4*len(params)+np.arange(5)*width*4.5,
               [f"{round(min(c), 2)}-{round(max(c), 2)}" for c in bins1])
    ax.set_xlabel(paramX)
    ax.set_ylabel(paramY)
    ax.set_title(f"{paramY} by {paramX} and {group}")

ax1 = fig.add_subplot(221)
sbsColumn(ax1, "species", "sepal_width", "sepal_length")
ax2 = fig.add_subplot(222)
sbsColumn(ax2, "species", "petal_width", "petal_length")
plt.show()
```
