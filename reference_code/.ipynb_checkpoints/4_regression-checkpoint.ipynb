### Muli-Variable Linear Regression with Train-Test Split:

```python
# train-test split
# set random state to get same results each time
# test size should not be too small to avoid overfitting, 
# but large enough that a representative set of data can be used
train, test = train_test_split(exampleData, test_size=0.3, random_state=4)

# define and print attributes of the model
# accepts multiple predictors
def printMultipleLinearRegression(predictors, response):
    # assign the predictor to x
    x = train.loc[:,predictors]
    x_test = test.loc[:,predictors]

    # assign the response to y
    y = train[response]
    y_test = test[response]

    # create a LinearRegression instance
    lr = LinearRegression()

    # fit the data using linear regression
    model = lr.fit(x,y)
    
    # score against the test group
    score = model.score(x_test, y_test)

    print("Predictors: ", predictors)
    print("Response: ", response)
    print()
    
    # print attributes and equation
    # code derived from lab 4
    print("The intercept of this model is:", round(model.intercept_, 4))
    print("The slope coefficient of this model is:", model.coef_)
    print("")
    print("Thus the equation is:", response, " =", model.intercept_.round(4), "+",
          " + ".join([" ".join(["(", str(coef.round(8)),"*",x.columns[i],")"]) for i,coef in enumerate(model.coef_)]))
    print("With an R-squared score of: ", score) # print r-squared score for comparison
    print()
    print()
    return model, score

# print example attributes of the model
print("=== Petal Length:")
lengthModel, lengthScore = printMultipleLinearRegression(["sepal_length"], "petal_length")
print("=== Petal Width:") # also print with memory as the response for completeness
widthModel, widthScore = printMultipleLinearRegression(["sepal_width"], "petal_width")

# make a prediction
values1 = [[6]]
print(lengthModel.predict(values1))
values2 = [[3]]
print(widthModel.predict(values2))
```

### Plot Linear Regression Model:

```python
# plot scatter points and regression line for comparison
def plotLinearRegression(ax, predictor, response, multiLinearModel, scoreMLR, baselineIndex):
    # get test values
    x_test = test[predictor]
    y_test = test[response]
    
    # scatter the points
    plt.scatter(x_test, y_test, alpha=0.5, c="green")

    # plot regression line
    b0M = multiLinearModel.intercept_
    b1M = multiLinearModel.coef_[baselineIndex]
    x_range = [x_test.min(), x_test.max()]                 # get the bounds for x
    y_rangeM = [b0M+b1M*x_range[0], b0M+b1M*x_range[1]]    # get the bounds for y
    plt.plot(x_range, y_rangeM, c="blue")

    plt.xlim(x_test.min(), x_test.max())
    plt.ylim(y_test.min(), y_test.max())
    
    # add some labels and titles
    ax.set_xlabel(f"{predictor}")
    ax.set_ylabel(f"{response}")
    ax.set_title(f"{response} by {predictor}")

# example plots
fig = plt.figure(figsize=(30,10))
fig.subplots_adjust(wspace=0.3, hspace=0.3)
ax1 = fig.add_subplot(221)
plotLinearRegression(ax1, "sepal_length", "petal_length", lengthModel, lengthScore, 0)
ax2 = fig.add_subplot(223)
plotLinearRegression(ax2, "sepal_width", "petal_width", widthModel, widthScore, 0) 
plt.show()
```
